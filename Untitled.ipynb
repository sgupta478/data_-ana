. Divide customers into bins based on tenure e.g. for tenure < 6 months: assign a tenure group if 1-6, for tenure between 1 to 1/2 Yrs, 

```python
# Get the max tenure
print(data_clean['tenure'].max()) #72
```

```python
# Create labels for the first two groups
first_two_labels = ['1-6', '7-12']

# Create labels for the remaining groups
rest_labels = ["{0} - {1}".format(i, i + 11) for i in range(13, 72, 12)]

# Combine all the labels
labels = first_two_labels + rest_labels

# Create bins for the first two groups
first_two_bins = [0, 6, 12]

# Create bins for the remaining groups
rest_bins = range(13, 80, 12)

# Combine all the bins
bins = first_two_bins + list(rest_bins[:-1])

# Use pd.cut() function to create the tenure_group column
data_clean['tenure_group'] = pd.cut(data_clean.tenure, bins=bins, labels=labels, include_lowest=True)
```

```python
data_clean['tenure_group'].value_counts()
```

```python
#drop column customerID and tenure
data_clean.drop(columns= ['customerID','tenure'], axis=1, inplace=True)
data_clean.head()
```

```python
def uniplot(df,col,title,hue =None):
    
    sns.set_style('whitegrid')
    sns.set_context('talk')
    plt.rcParams["axes.labelsize"] = 20
    plt.rcParams['axes.titlesize'] = 22
    plt.rcParams['axes.titlepad'] = 30
    
    
    temp = pd.Series(data = hue)
    fig, ax = plt.subplots()
    width = len(df[col].unique()) + 7 + 4*len(temp.unique())
    fig.set_size_inches(width , 8)
    plt.xticks(rotation=45)
    plt.yscale('log')
    plt.title(title)
    ax = sns.countplot(data = df, x= col, order=df[col].value_counts().index,hue = hue,palette='bright') 
        
    plt.show()
```

```python
uniplot(new_df1_target1,col='Partner',title='Distribution of Gender for Churned Customers',hue='gender')
```
